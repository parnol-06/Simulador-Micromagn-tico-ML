```python
import micromagneticmodel as mm
import discretisedfield as df
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mp
import oommfc as oc
import random

Lx = 114e-9  # Doble del tamaño para acomodar dos esferas
Ly = 42e-9  # Mantenemos el mismo tamaño
Lz = 42e-9  # Mantenemos el mismo tamaño
c = 3e-9
# Después de la línea c = 3e-9
#radius_squared = float(input("Ingrese el radio de las esferas en nm: ")) * 1e-9  # Convertir a metros
#distancia_borde = abs(float(input("Ingrese la distancia desde el borde de la esfera al origen en nm: ")) * 1e-9)  # Convertir a metros y tomar valor absoluto


# Cálculo de la separación entre los bordes
#separacion_bordes = max(0, 2 * distancia_borde )
#print(f"Separación entre los bordes de las nanoesferas: {separacion_bordes} metros")

#print(f"Separación entre los bordes de las nanoesferas: {separacion_bordes} metros")
#radius_squared = float(input("Ingrese el radio de las esferas en nm: ")) * 1e-9  # Convertir a metros
#E_0 = float(input("Ingrese la distancia entre el borde de la esfera y el origen en nm: ")) * 1e-9  # Convertir a metros
# Definimos los puntos de la región
p1 = (-Lx/2, -Ly/2, -Lz/2)
p2 = (Lx/2, Ly/2, Lz/2)
cell = (c, c, c)

# Definimos la región y la malla
region = df.Region(p1=p1, p2=p2)
mesh = df.Mesh(region=region, cell=cell)

# Visualización de la malla
mesh.mpl()
mesh.k3d()

# Definimos el sistema
system = mm.System(name="ensayo4")
gamma0 = mm.consts.gamma0
alpha = 1
system.dynamics = mm.Precession(gamma0=gamma0) + mm.Damping(alpha=alpha)

# Parámetros del material, todos en el S.I
Ms = 1.7e6  # Magnetización de saturación (A/m)
K = 48e3    # Constante de anisotropía (J/m)
A = 2.1e-11 # Constante de intercambio
H = (0/mm.consts.mu0, 0, 0)  # Campo aplicado, se divide por la permeabilidad magnética con el fin de expresarlo en el S.I
u1 = (1, 0, 0)
u2 = (0, 1, 0)
u3 = (0, 0, 1)
#separation between spheres = 2*((r1_squared)-(r1_squared-(radius_squared))
#######################################################################################################################
#########################################################################################################################
#MODIICA AQUÍ:
radius_squared = 21e-9   ###############RADIO DE LAS ESFERAS
E_0 = 36e-9 ############################DISTANCIA AL ORIGEN DE LOS BORDES DE LAS ESFERAS
# Función para definir las dos esferas
def Ms_fun(point):
    x, y, z = point
    # Centro de la primera esfera en (-Lx/4, 0, 0)
    #r1_squared = (x + Lx/3)**2 + y**2 + z**2
    #edge of sphere to origin E_0
    
    r1_squared = (x + E_0)**2 + y**2 + z**2
    # Centro de la segunda esfera en (Lx/4, 0, 0)
    #r2_squared = (x - Lx/3)**2 + y**2 + z**2
    r2_squared = (x - E_0)**2 + y**2 + z**2
    #radius_squared = (Lx/4)**2  # Radio de las esferas
    radius_squared = (21e-9)**2  # Radio de las esferas##########################################################MODIFICA AQUI:
    if r1_squared <= radius_squared or r2_squared <= radius_squared:
        return Ms
    else:
        return 0


#separation between spheres (S)
# Separación entre esferas (S)
#radius_squared = (21e-9)**2
#E_0 = 30e-9   ###########################3#######################################################################MODIFICA AQUI:
S = 2 * (E_0 - radius_squared )  # Convertir a metros
print(f"Separación entre esferas (S): {S} metros")

# Energía del sistema
system.energy = (
    mm.Exchange(A=A) +
    mm.CubicAnisotropy(K=K, u1=u1, u2=u2) +
    mm.Zeeman(H=H) +
    mm.Demag()
)

# Campo magnético inicial aleatorio
#system.m = df.Field(mesh, nvdim=3, value=lambda point: [2*random.random()-1 for i in range(3)], norm=Ms_fun)
system.m = df.Field(mesh, nvdim=3, value=(1,0,0), norm=Ms_fun)


# Visualización de la orientación en z
system.m.orientation.sel("z").mpl()

# Visualización en 3D
plot = k3d.plot()

# Para visualizar la región
for z in np.linspace(p1[2], p2[2], 10):
    plot += k3d.line(
        [
            [p1[0], p1[1], z],
            [p2[0], p1[1], z],
            [p2[0], p2[1], z],
            [p1[0], p2[1], z],
            [p1[0], p1[1], z]
        ],
        shader="flat",
        color=0x0000ff,
    )

# Para visualizar las esferas
points = []
for x in np.linspace(p1[0], p2[0], 40):
    for y in np.linspace(p1[1], p2[1], 40):
        for z in np.linspace(p1[2], p2[2], 40):
            if Ms_fun((x, y, z)) > 0:
                points.append([x, y, z])

plot += k3d.points(points, point_size=0.3, color=0xff0000)

plot.display()
```
